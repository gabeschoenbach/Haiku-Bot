```python
def findHaikus(dataset):
    
    data = dataset

    idx = 0
    stick = 0
    line = [0, None, None]

    haiku = [[], [], []]

    haikuList = []

    while (idx < len(data)):
    #         print(idx, ":", data[idx])
        if (line[0] is not None):
            if (line[0] == 0):
                stick = idx
            if (data[idx] == -1):
#                 print("Found a -1...")
                haiku = [[],[],[]]
                line = [0, None, None]
                idx = stick + 1
                continue
            line[0] += data[idx]
            if (line[0] == 5):
#                 print("Found first line")
                for i in range(stick, idx+1):
                    haiku[0].append(i)
                line = [None, 0, None]
            elif (line[0] > 5):
    #                 print("Overshot first line")
                haiku = [[],[],[]]
                line = [0, None, None]
                idx = stick + 1
                continue
        elif (line[1] is not None):
            if (line[1] == 0):
                stick1 = idx
            if (data[idx] == -1):
                haiku = [[],[],[]]
                line = [0, None, None]
                idx = stick + 1
                continue
            line[1] += data[idx]
            if (line[1] == 7):
    #                 print("Found second line")
                for i in range(stick1, idx+1):
                    haiku[1].append(i)
                line = [None, None, 0]
            elif (line[1] > 7):
    #                 print("Overshot second line")
                haiku = [[],[],[]]
                line = [0, None, None]
                idx = stick + 1
                continue
        elif (line[2] is not None):
            if (line[2] == 0):
                stick2 = idx
            if (data[idx] == -1):
                haiku = [[],[],[]]
                line = [0, None, None]
                idx = stick + 1
                continue
            line[2] += data[idx]
            if (line[2] == 5):
    #             print("Found Haiku! (starting at index", stick, ")")
                for i in range(stick2, idx+1):
                    haiku[2].append(i)
    #                 print(haiku)
                haikuList.append(haiku)
                haiku = [[],[],[]]
                line = [0, None, None]
                idx = stick + 1
                continue
            elif (line[2] > 5):
    #                 print("Overshot third line")
                haiku = [[],[],[]]
                line = [0, None, None]
                idx = stick + 1
                continue
        idx += 1

    numHaikus = len(haikuList)
    pct = numHaikus / len(data)
#     print("The % of haikus is:", pct)
    return haikuList
```
